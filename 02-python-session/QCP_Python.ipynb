### The `print()` Function in Python

The `print()` function is a fundamental built-in function in Python used to display output to the console. It can show various types of data, including:

*   **Text (strings)**: Enclosed in single or double quotes, e.g., `print("Hello, World!")`.
*   **Numbers**: Integers, floats, etc., e.g., `print(42)` or `print(3.14)`.
*   **Variables**: The current value stored in a variable, e.g., `x = 10; print(x)`.
*   **Expressions**: The result of an arithmetic or logical operation, e.g., `print(5 + 7)`.

**Key Characteristics:**

*   **Multiple Arguments**: You can pass multiple items to `print()` by separating them with commas. By default, `print()` will insert a space between each item.
    *   Example: `print("Name:", name, "Age:", age)`
*   **Automatic Newline**: By default, `print()` adds a newline character at the end of its output, so subsequent `print()` calls will start on a new line. This behavior can be changed using the `end` parameter (e.g., `end=' '`).
*   **F-strings (Formatted String Literals)**: A modern and highly readable way to embed variable values or expressions directly within a string. They are prefixed with an `f` or `F`.
    *   Example: `print(f"x is equal to {x}")`

````python
x = input("input x")

x = 42
print("x is equal to", x)
print(f"x is equal to {x}")
print("\n" + "-"*30 + "\n")

a = 5
b = 7
print("sum =", a + b)
print(f"sum = {a + b}")
print("\n" + "-"*30 + "\n")

name = "ali"
age = 21
print("Name:", name, "Age:", age)
print(f"Name" + ":", name, "Age:", age)
````

### **Explanation of Arithmetic Operators**

Arithmetic operators are used to perform common mathematical calculations on numerical values. Python supports several such operators:

*   **Addition (`+`)**: Adds two operands.
*   **Subtraction (`-`)**: Subtracts the second operand from the first.
*   **Multiplication (`*`)**: Multiplies two operands.
*   **Exponentiation (`**`)**: Raises the first operand to the power of the second (e.g., `a**b` means `a` to the power of `b`).
*   **Division (`/`)**: Divides the first operand by the second. The result is always a float.
*   **Floor Division (`//`)**: Divides the first operand by the second, but the result is floored to the nearest whole number (integer division). The result is an integer.
*   **Modulo (`%`)**: Returns the remainder when the first operand is divided by the second.

Let's look at the example in the notebook to understand these operations better, where `a = 13` and `b = 5`.

````python
a = 13
b = 5
print("a =",a)
print("b =",b)
print()

print("a + b =", a+b)
print("a - b =", a-b)
print("a * b =", a*b)
print("a ^ b =", a**b)
print("a / b =", a/b)
print("a // b = " , a//b)
print("a mod b =", a % b)
````

### **What are Variables?**

In programming, a **variable** is a named storage location that holds a value. Think of it as a container with a label (the variable name) where you can store different types of information. Key characteristics of variables in Python:

A **string** in Python is an ordered, immutable sequence of characters. This means:

*   **Ordered**: Characters maintain their position. You can access individual characters or substrings by their index.
*   **Immutable**: You cannot change individual characters within a string after it's created. Any operation that seems to 'modify' a string actually creates a new string.

### **What is a List?**

A **list** in Python is an ordered, mutable collection of elements. This means:

*   **Ordered**: Elements maintain their insertion order. You can access elements by their index.
*   **Mutable**: You can change, add, or remove elements from a list after it's created.
*   **Allows Duplicates**: A list can contain multiple instances of the same element.

````python
list1 = [10, 20, 20, 30, 40, 50, 60]
print(list1)
print(type(list1))       # -> list

print(list1[3])          # -> 30
print(list1[-2])         # -> 50

list1[0] = 90
print(list1)             # -> [90, 20, 20, 30, 40, 50, 60]

list1.append(70)
print(list1)             # -> [90, 20, 20, 30, 40, 50, 60, 70]

list1.insert(2, 100)
print(list1)             # -> [90, 20, 100, 20, 30, 40, 50, 60, 70]

list1.sort()
print(list1)             # -> [20, 20, 30, 40, 50, 60, 70, 90, 100]

list1.remove(100)
print(list1)             # -> [20, 20, 30, 40, 50, 60, 70, 90]

list1.pop(3)
print(list1)             # -> [20, 20, 30, 50, 60, 70, 90]

list1.pop()
print(list1)             # -> [20, 20, 30, 50, 60, 70]

print(list1.index(30))   # -> 2
print(list1.count(20))   # -> 2

list1.reverse()
print(list1)             # -> [70, 60, 50, 30, 20, 20]
````

### **What is a Tuple?**

A **tuple** in Python is an ordered, immutable collection of elements. This means:

*   **Ordered**: Elements maintain their insertion order. You can access elements by their index.
*   **Immutable**: You cannot change, add, or remove elements from a tuple after it's created. Once defined, its contents are fixed.
*   **Allows Duplicates**: A tuple can contain multiple instances of the same element.

````python
tuple1 = ("Spin","Hadamard", "Qubit", "Superposition", "Entanglement", "Schrodinger")
print(tuple1)

print(type(tuple1))               # tuple

print(tuple1[2])                  # -> Qubit
print(len(tuple1))                # -> 6

list2 = list(tuple1)
print(list2)                      # -> ['Spin', 'Hadamard', 'Qubit', 'Superposition', 'Entanglement', 'Schrodinger']

(x,y,*z) = tuple1
print(x)                          # -> Spin
print(y)                          # -> Hadamard
print(z)                          # -> ['Qubit', 'Superposition', 'Entanglement', 'Schrodinger', 'Quantum Physics']
print(type(z))                    # -> list

tuple1 += ("Quantum Physics",)
print(tuple1)                     # -> ('Spin', 'Hadamard', 'Qubit', 'Superposition', 'Entanglement', 'Schrodinger', 'Quantum Physics')
````

### **Conditional Logic: `if`, `elif`, and `else`**

In Python, `if`, `elif` (else if), and `else` statements form the backbone of conditional execution, allowing your program to make decisions based on various conditions.

*   **`if`**: The initial condition. If `True`, its block is executed. If `False`, Python checks the `elif` conditions.
*   **`elif`**: Stands for "else if". These are checked sequentially if the preceding `if` or `elif` conditions were `False`. The first `elif` condition that evaluates to `True` will have its block executed, and the rest of the chain is skipped.
*   **`else`**: An optional final block that executes if *all* preceding `if` and `elif` conditions are `False`.

**Key Principle**: Only *one* block of code within an `if-elif-else` chain will ever execute.

````python
grade = 75

print(f"Checking grade: {grade}\n")

if grade >= 90:
    print("Excellent! You received an A.")
elif grade >= 80:
    print("Very good! You received a B.")
elif grade >= 70:
    print("Good job! You received a C.")
elif grade >= 60:
    print("You passed! You received a D.")
else:
    print("Unfortunately, you did not pass. You received an F.")

print("\nGrade assessment complete.")
````

### **Introduction to `while` Loops**

A `while` loop in Python is used to execute a block of code repeatedly as long as a certain condition remains `True`. It's ideal when you don't know beforehand how many times you need to iterate, and the loop should continue based on a condition.

### How `while` Loops Work:

1.  **Condition Check**: The loop first evaluates a given `condition`.
2.  **Execution**: If the `condition` is `True`, the code inside the loop's body is executed.
3.  **Repeat**: After executing the body, the `condition` is checked again. This process repeats until the `condition` becomes `False`.
4.  **Termination**: When the `condition` is `False`, the loop terminates, and the program continues with the code immediately following the loop.

### Basic Syntax:

```python
while condition:
    # Code to be executed repeatedly as long as the condition is True
    statement_1
    statement_2
    # ...
```

**Key Points:**

*   **`while` keyword**: Marks the beginning of the loop.
*   **`condition`**: An expression that evaluates to `True` or `False`. This condition *must* eventually become `False` for the loop to terminate. If it never becomes `False`, you'll have an **infinite loop**.
*   **Colon (`:`)**: Required at the end of the `while` line.
*   **Indentation**: The code block inside the `while` loop **must be indented**.

### Important Consideration: Avoiding Infinite Loops

It's crucial to ensure that the `condition` within a `while` loop will eventually become `False`. This typically involves modifying one or more variables within the loop's body that are part of the condition. Forgetting to do so will result in an **infinite loop**, where the program keeps running indefinitely, potentially consuming all system resources. You might need to manually stop the execution (e.g., by pressing the stop button in your environment).

Let's look at some examples to understand `while` loops better.

````python
number = 9
factorial = 1
original_number = number

print(f"Calculating factorial for {number}:\n")

if number < 0:
    print("Factorial is not defined for negative numbers.")
elif number == 0:
    print("The factorial of 0 is 1.")
else:
    while number > 0:
        factorial *= number
        number -= 1
    print(f"The factorial of {original_number} is {factorial}.")
````

### **Introduction to `for` Loops**

A `for` loop in Python is used for iterating over a sequence (that is, a list, a tuple, a dictionary, a set, or a string) or other iterable objects. It allows you to execute a block of code once for each item in the sequence.

### How `for` Loops Work:

1.  **Iteration**: The `for` loop iterates through each item in a given sequence.
2.  **Assignment**: In each iteration, the current item from the sequence is assigned to a loop variable.
3.  **Execution**: The code block inside the loop is executed with the current item.
4.  **Completion**: The loop continues until all items in the sequence have been processed.

### Basic Syntax:

```python
for item in sequence:
    # Code to be executed for each item
    statement_1
    statement_2
    # ...
```

**Key Points:**

*   **`for` keyword**: Marks the beginning of the loop.
*   **`item`**: A temporary variable that takes on the value of each element in the `sequence` during each iteration.
*   **`in` keyword**: Specifies the sequence you are iterating over.
*   **`sequence`**: Any iterable Python object (list, tuple, string, range, dictionary, set).
*   **Colon (`:`)**: Required at the end of the `for` line.
*   **Indentation**: The code block inside the `for` loop **must be indented**.

### Common Uses of `for` Loops:

*   Processing elements in a list, tuple, or string.
*   Repeating an action a specific number of times using `range()`.
*   Iterating through dictionary keys, values, or items.

Let's look at some examples to understand `for` loops better.

````python
import math
from math import pi

I0 = 100.0  # arbitrary units

print(f"Initial light intensity (I0): {I0}\n")
print("Calculating intensity for different angles:\n")

for angle_degrees in range(0, 181, 30):
    angle_radians = math.radians(angle_degrees)

    intensity = I0 * (math.cos(angle_radians) ** 2)
    print(f"Angle: {angle_degrees}° | Intensity (I): {intensity:.2f}")
````

**Libraries** (or modules/packages) in Python are pre-written blocks of code that extend your program's functionality. They promote code reusability, modularity, and efficiency, offering specialized tools for various tasks.

**Common ways to import:**

1.  **Standard Import (`import library_name`)**: Imports the entire library. Access components via `library_name.component`.
2.  **Import with Alias (`import library_name as alias_name`)**: Imports the library and assigns it a shorter alias (e.g., `import numpy as np`).

**Best Practices:**

*   Place imports at the top of your script/notebook.
*   Use conventional aliases for common libraries.

### What are Functions in Python?

A **function** in Python is a block of organized, reusable code that is used to perform a single, related action. Functions provide better modularity for your application and a high degree of code reusing.

**Key Concepts:**

*   **Definition**: Functions are defined using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. Parameters (inputs) can be placed inside the parentheses.
*   **Docstrings**: A string literal immediately after the function header can serve as a documentation string (docstring) to explain what the function does.
*   **Body**: The code block within every function starts with an indentation.
*   **Return Statement**: The `return` statement is used to exit a function and pass back a value (or values) to the caller. If no `return` statement is used, the function implicitly returns `None`.
*   **Calling a Function**: To execute the code inside a function, you 'call' it by its name followed by parentheses, potentially passing arguments if the function expects parameters.

**Why use functions?**

1.  **Modularity**: Breaking down complex problems into smaller, manageable sub-problems makes the code easier to understand and maintain.
2.  **Reusability**: Once defined, a function can be called multiple times from different parts of your program, avoiding redundant code.
3.  **Readability**: Functions help organize code logically, making it easier for others (and your future self) to read and comprehend.
4.  **Debugging**: Isolating code into functions makes it easier to pinpoint and fix errors.

**Basic Syntax:**

```python
def function_name(parameter1, parameter2):
    """Docstring: Explanation of what the function does."""
    # Function body - code to be executed
    result = parameter1 + parameter2
    return result

# Calling the function
output = function_name(10, 20)
print(output)
```

Let's see an example of a simple function in action.

````python
def greet(name):

    return f"Hello, {name}!"

print(greet("Asja"))        # → Hello, Asja!
print(greet("Ali"))         # → Hello, Ali!



def greet(name="my friend", language="English"):
    if language == "English":
        print(f"Hello, {name}!")
    elif language == "Persian":
        print(f"Salam، {name}!")
    elif language == "French":
        print(f"Bonjour, {name}!")
    elif language == "German":
        print(f"Hallo, {name}!")

greet()
greet("Marie")
greet("Sara", "Persian")
greet(language="French", name="Luc")
greet(language="German", name="Armin")
````

### What is NumPy?

**NumPy** (Numerical Python) is a fundamental library for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays. NumPy is the backbone of many other scientific and data analysis libraries in Python, including Pandas, SciPy, and Matplotlib.

**Key Features and Concepts:**

*   **`ndarray` (N-dimensional Array Object)**: The core object in NumPy is the `ndarray`, which is a fast, flexible container for large datasets in Python. Arrays are much more efficient than Python's built-in lists for numerical operations, especially with large amounts of data.
*   **Vectorized Operations**: NumPy allows you to perform operations on entire arrays without writing explicit loops. This 'vectorization' significantly speeds up computations, as many operations are implemented in C or Fortran.
*   **Broadcasting**: A powerful mechanism that allows NumPy to work with arrays of different shapes when performing arithmetic operations. Subject to certain constraints, the smaller array is 'broadcast' across the larger array so that they have compatible shapes.
*   **Mathematical Functions**: NumPy provides a vast collection of mathematical functions for linear algebra, Fourier transforms, random number generation, and more.

**Why use NumPy?**

*   **Performance**: Operations on NumPy arrays are often orders of magnitude faster than equivalent operations on standard Python lists, especially for large datasets.
*   **Memory Efficiency**: NumPy arrays consume less memory than Python lists for storing numerical data.
*   **Convenience**: Its rich set of functions and operators makes it easy to perform complex mathematical computations with concise code.
*   **Foundation for Data Science**: It's a foundational library that almost every other data science library in Python builds upon.

Let's look at some basic examples of how to create and manipulate NumPy arrays.

````python
import numpy as np

print("--- Array Creation ---")

# Creating an array from a Python list
arr1 = np.array([1, 2, 3, 4, 5])
print(f"Array from list: {arr1}")
print(f"Shape of arr1: {arr1.shape}")
print(f"Type of elements in arr1: {arr1.dtype}")

# Creating a 2D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\n2D Array:\n{arr2}")
print(f"Shape of arr2: {arr2.shape}")

# Creating arrays with zeros or ones
zeros_arr = np.zeros((2, 3))
print(f"\nArray of zeros:\n{zeros_arr}")

ones_arr = np.ones((3, 2))
print(f"\nArray of ones:\n{ones_arr}")

# Creating a sequence of numbers
range_arr = np.arange(0, 10, 2) # start, stop (exclusive), step
print(f"\nArray from arange: {range_arr}")

# Creating a linearly spaced array
linspace_arr = np.linspace(0, 1, 5) # start, stop (inclusive), number of elements
print(f"\nArray from linspace: {linspace_arr}")

print("\n--- Basic Array Operations ---")

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

# Element-wise addition
print(f"Addition (a + b): {a + b}")

# Element-wise multiplication
print(f"Multiplication (a * b): {a * b}")

# Scalar multiplication
print(f"Scalar multiplication (a * 2): {a * 2}")

# Dot product (matrix multiplication for 1D arrays, or specific for 2D)
# For 1D arrays, this is the inner product (sum of products)
print(f"Dot product (a . b): {np.dot(a, b)}")

print("\n--- Universal Functions (ufuncs) ---")
c = np.array([1, 4, 9, 16])
print(f"Square root of c: {np.sqrt(c)}")

print(f"Exponential of a: {np.exp(a)}")

print("\n--- Aggregation Functions ---")
rand_arr = np.array([5, 2, 8, 1, 9, 3])
print(f"Original array: {rand_arr}")
print(f"Sum of elements: {np.sum(rand_arr)}")
print(f"Mean of elements: {np.mean(rand_arr)}")
print(f"Maximum element: {np.max(rand_arr)}")
print(f"Minimum element: {np.min(rand_arr)}")
print(f"Standard deviation: {np.std(rand_arr)}")

# Aggregation along axes for 2D arrays
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\nMatrix:\n{matrix}")
print(f"Sum along columns (axis=0): {np.sum(matrix, axis=0)}")
print(f"Sum along rows (axis=1): {np.sum(matrix, axis=1)}")
````
